```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar el dataset
df = sns.load_dataset('tips')

# Seleccionar solo las columnas numricas
df_numeric = df.select_dtypes(include=['int64', 'float64'])

# Anlisis de correlacin y mapa de calor
corr = df_numeric.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True)
plt.title('Mapa de Calor de Correlacin')
plt.show()

# Subplots para comparar diferentes variables clave
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
sns.scatterplot(x='total_bill', y='tip', data=df, ax=axs[0, 0])
axs[0, 0].set_title('Relacin entre Total Bill y Tip')
sns.scatterplot(x='total_bill', y='size', data=df, ax=axs[0, 1])
axs[0, 1].set_title('Relacin entre Total Bill y Tamao del Grupo')
sns.scatterplot(x='tip', y='size', data=df, ax=axs[1, 0])
axs[1, 0].set_title('Relacin entre Tip y Tamao del Grupo')
sns.scatterplot(x='total_bill', y='time', data=df, ax=axs[1, 1])
axs[1, 1].set_title('Relacin entre Total Bill y Hora del Da')
plt.tight_layout()
plt.show()

# Aadir cuadrculas, leyendas, anotaciones y flechas a los grficos
fig, ax = plt.subplots(figsize=(12, 6))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='time', ax=ax)
ax.set_title('Relacin entre Total Bill y Tip con Anotaciones y Flechas')
ax.set_xlabel('Total Bill ($)')
ax.set_ylabel('Tip ($)')
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')
ax.legend(loc='upper right', fontsize='large', shadow=True, title='Detalles')
ax.annotate('Mayor tip en la cena', xy=(48.27, 9.0), xytext=(50, 10), arrowprops=dict(facecolor='black', shrink=0.05))
ax.arrow(30, 5, 15, 4, head_width=0.5, head_length=0.5, fc='green', ec='green')
plt.show()
```
